Adding columns where the respondent has not answered the multiple choice question.

```python
ques = np.array(ks1.columns)
q_list = [re.sub(r"_\d+","",x) for x in ques]
unique, counts = np.unique(q_list, return_counts=True)
q_arr = (np.asarray((unique, counts)).T)
q_count = pd.DataFrame(q_arr, columns = ['question','count'])
q_count['count'] = q_count['count'].astype('int16')
# ques_dict = dict(q_count[q_count['count'] != 1].values)
ques_dict = dict(q_count.values)
ques_dict
```

```python
for key,value in tqdm(ques_dict.items()):
    cols = [col for col in ks1.columns if f'{key}' in col]
    ks1.loc[ks1[cols].isnull().sum(axis=1)==value,f'{key}_X'] = 'No Answer'
```
